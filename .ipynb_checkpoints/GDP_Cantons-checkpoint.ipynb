Quick EDA and Data Visualization on Swiss Cantons GDP data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('GDP_Cantons.csv')
df.head()
```

```python
df=df.drop(0)
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
# Unpivot the columns
df=df.melt(id_vars =['Canton'], var_name = 'Year', value_name = 'GDP')
df
```

```python
# Let's inspect the datatype
df.info()
```

```python
# Convert the GDP object datatype to integer
df["GDP"]=df["GDP"].str.replace("’","", regex=True)
df["GDP"]=df["GDP"].astype(int)
df
```

```python
# Convert year column to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.year
```

```python
df['GDP_Billions'] = round(df['GDP'] / 1_000)
df.drop(columns="GDP",inplace=True)
```

```python
# Select top 10 swiss cantons by gdp
df[(df['Year'] == 2019) & (df["Canton"]!="Switzerland")].sort_values(by='GDP_Billions')[-10:].reset_index()
```

```python
# List the top 10 cantons to use it in the graph
cantons = df[(df['Year'] == 2019) & (df["Canton"]!="Switzerland")].sort_values(by='GDP_Billions')[-10:]['Canton'].values
cantons
```

```python
# Setup plot size.
fig, ax = plt.subplots(figsize=(7,4))

# Create grid 
ax.grid(which="major", axis='both', color='#758D99', alpha=0.6, zorder=1)

# Remove splines.
ax.spines[['top','right','bottom']].set_visible(False)

# Setup data
gdp_dbell = (df[(df['Canton'].isin(cantons)) & ((df['Year'] == 2008) | (df['Year'] == 2019))][['Year','GDP_Billions','Canton']]
                .pivot(index='Canton',columns='Year', values='GDP_Billions')
                .sort_values(by=2019))

# Plot data
# Plot horizontal lines first
ax.hlines(y=gdp_dbell.index, xmin=gdp_dbell[2008], xmax=gdp_dbell[2019], color='#758D99', zorder=2, linewidth=2, label='_nolegend_', alpha=.8)
# Plot bubbles next
ax.scatter(gdp_dbell[2008], gdp_dbell.index, label='2008', s=60, color='#DB444B', zorder=3)
ax.scatter(gdp_dbell[2019], gdp_dbell.index, label='2019', s=60, color='#006BA2', zorder=3)

# Set xlim
ax.set_xlim(0, 160)

# Reformat x-axis tick labels
ax.xaxis.set_tick_params(labeltop=True,      
                         labelbottom=False,  
                         bottom=False,       
                         labelsize=11,       
                         pad=-1)            

# Reformat y-axis tick labels
ax.set_yticklabels(gdp_dbell.index,       
                   ha = 'left')             
ax.yaxis.set_tick_params(pad=100,            
                         labelsize=11,      
                         bottom=False)       


# Reformat x-axis tick labels
ax.xaxis.set_tick_params(labeltop=True,      
                         labelbottom=False,  
                         bottom=False,      
                         labelsize=11,      
                         pad=-1)             

# Reformat y-axis tick labels
ax.set_yticklabels(gdp_dbell.index,       
                   ha = 'left')              
ax.yaxis.set_tick_params(pad=80,            
                         labelsize=11,       
                         bottom=False)       

# Set Legend
ax.legend(['2008', '2019'], loc=(-.29,1.09), ncol=2, frameon=False, handletextpad=-.1, handleheight=1)

# Add in line and tag
ax.plot([-0.08, .9],                
        [1.17, 1.17],                
        transform=fig.transFigure,  
        clip_on=False, 
        color='#E3120B', 
        linewidth=.6)
ax.add_patch(plt.Rectangle((-0.08,1.17),             
                           0.05,                      
                           -0.025,                      
                           facecolor='#E3120B', 
                           transform=fig.transFigure, 
                           clip_on=False, 
                           linewidth = 0))

# Add in title and subtitle
ax.text(x=-0.08, y=1.09, s="Top 10 Swiss Cantons by GDP", transform=fig.transFigure, ha='left', fontsize=13, weight='bold', alpha=.8)
ax.text(x=-0.08, y=1.04, s="In billions of CHF", transform=fig.transFigure, ha='left', fontsize=11, alpha=.8)

# Set source text
ax.text(x=-0.08, y=0.04, s="""Source: "GDP by Canton" via Swiss Federal Statistical Office""", transform=fig.transFigure, ha='left', fontsize=9, alpha=.7)
        
```
