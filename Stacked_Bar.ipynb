```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format ='retina'
```

```python
df=pd.read_csv("p1_communes.csv",index_col=0)
```

```python
list_canton=df.index.unique().dropna
print(list_canton)
```

```python
df=df.dropna()
df.head(10)
```

```python
y_cor=np.arange(10)
```

```python
# Let's select the top 10 most populated communes and the area columns we need for our chart.
df2=df[["Commune","Settlement area","Agricultural area","Wooded area","Unproductive area","Residents"]].sort_values("Residents",ascending=False).head(10)
df2
```

```python
# List of top 10 communes
communes_pop_plus =df2.Commune
communes_pop_plus
```

```python
# Drop the 'Residents' column as we don't need it
df3=df2.drop("Residents",axis=1).set_index('Commune')
df3
```

```python
communes=df2["Commune"].to_list()
wooded_area=np.array(df2["Wooded area"].to_list())
agricultural_area=np.array(df2["Agricultural area"].to_list())
settlement_area=np.array(df2["Settlement area"].to_list())
unproductive_area=np.array(df2["Unproductive area"].to_list())
total= wooded_area+agricultural_area+settlement_area+unproductive_area
```

```python
#I use here plt.barh to be able to do the stacked bar. 

plt.barh(y_cor,settlement_area)
plt.barh(y_cor,agricultural_area, left=settlement_area)
plt.barh(y_cor,wooded_area, left=settlement_area + agricultural_area)
plt.barh(y_cor,unproductive_area, left=settlement_area + agricultural_area + wooded_area)
plt.xlabel("Area type in %")
plt.ylabel("Commmune")
plt.title("Area type for top 10 swiss communes")
plt.legend(labels=["Settlement area","Agricultural area","Wooded area","Unproductive area"],loc="upper left",bbox_to_anchor=(1.02,1))
plt.yticks(y_cor,communes)
plt.style.use('fivethirtyeight')


# Export plot as high resolution PNG
plt.savefig('plt_stacked.png',    # Set path and filename
            dpi = 300,                     # Set dots per inch
            bbox_inches="tight")           # Remove extra whitespace around plot
```
